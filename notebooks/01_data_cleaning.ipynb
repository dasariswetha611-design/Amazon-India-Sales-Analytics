```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
files = glob.glob("../data/raw/amazon_india_*.csv")

transaction_files = [
    file for file in files
    if "products_catalog" not in file
]

df_list = []

for file in transaction_files:
    df = pd.read_csv(file)
    df_list.append(df)

transactions = pd.concat(df_list, ignore_index=True)

print(transactions.shape)
transactions.head()
```

```python
#Challenge 1: Date Standardization
transactions["order_date"] = pd.to_datetime(
    transactions["order_date"],
    errors="coerce"
)

transactions["order_date"].isna().sum()
```

```python
#Challenge 2: Price Cleaning
price_cols = [
    "original_price_inr",
    "discounted_price_inr",
    "subtotal_inr",
    "final_amount_inr"
]

for col in price_cols:
    if col in transactions.columns:
        transactions[col] = (
            transactions[col]
            .astype(str)
            .str.replace("₹", "", regex=False)
            .str.replace(",", "", regex=False)
            .str.replace("Price on Request", "", regex=False)
            .str.strip()
        )

        transactions[col] = pd.to_numeric(
            transactions[col],
            errors="coerce"
        )

        transactions[col] = transactions[col].fillna(
            transactions[col].median()
        )
```

```python
#Challenge 3: Customer Rating Standardization
transactions["customer_rating"] = (
    transactions["customer_rating"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")[0]
)

transactions["customer_rating"] = pd.to_numeric(
    transactions["customer_rating"],
    errors="coerce"
)

transactions.loc[
    (transactions["customer_rating"] < 1) |
    (transactions["customer_rating"] > 5),
    "customer_rating"
] = np.nan

transactions["customer_rating"] = transactions["customer_rating"].fillna(
    transactions["customer_rating"].median()
)
```

```python
#Challenge 4: City Standardization
transactions["customer_city"] = (
    transactions["customer_city"]
    .astype(str)
    .str.strip()
    .str.title()
)

city_map = {
    "Bombay": "Mumbai",
    "Bangalore": "Bengaluru",
    "New Delhi": "Delhi",
    "Delihi": "Delhi",
    "Hydrabad": "Hyderabad",
    "Chennaii": "Chennai",
    "Kolkatta": "Kolkata"
}

transactions["customer_city"] = transactions["customer_city"].replace(city_map)
```

```python
#Challenge 5: Boolean Standardization

bool_map = {
    "Yes": True,
    "No": False,
    "Y": True,
    "N": False,
    "TRUE": True,
    "FALSE": False,
    "True": True,
    "False": False,
    "1": True,
    "0": False,
    1: True,
    0: False
}

for col in ["is_prime_member", "is_festival_sale"]:
    if col in transactions.columns:
        transactions[col] = transactions[col].replace(bool_map)
```

```python
# Challenge 6: Category Standardization

transactions["category"] = (
    transactions["category"]
    .astype(str)
    .str.strip()
    .str.title()
)

category_map = {
    "Electronic": "Electronics",
    "Electronics & Accessories": "Electronics",
    "Electronics Accessories": "Electronics",
    "Fashion & Apparel": "Fashion",
    "Home And Kitchen": "Home & Kitchen"
}

transactions["category"] = transactions["category"].replace(category_map)

transactions["category"].value_counts()
```

```python
# Challenge 7: Delivery Days Cleaning

transactions["delivery_days"] = (
    transactions["delivery_days"]
    .astype(str)
    .str.extract(r"(\d+)")[0]
)

transactions["delivery_days"] = pd.to_numeric(
    transactions["delivery_days"],
    errors="coerce"
)

transactions.loc[
    (transactions["delivery_days"] < 0) |
    (transactions["delivery_days"] > 30),
    "delivery_days"
] = np.nan

transactions["delivery_days"] = transactions["delivery_days"].fillna(
    transactions["delivery_days"].median()
)

transactions["delivery_days"].describe()
```

```python
# Challenge 8: Duplicate Transaction Handling

duplicates_before = transactions.duplicated().sum()

transactions = transactions.drop_duplicates()

duplicates_after = transactions.duplicated().sum()

print("Before:", duplicates_before)
print("After:", duplicates_after)
```

```python
# Challenge 9: Price Outlier Detection

Q1 = transactions["final_amount_inr"].quantile(0.25)
Q3 = transactions["final_amount_inr"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

transactions["price_outlier_flag"] = (
    (transactions["final_amount_inr"] < lower_limit) |
    (transactions["final_amount_inr"] > upper_limit)
)

print(
    transactions["price_outlier_flag"].value_counts()
)
```

```python
# Challenge 10: Payment Method Standardization

transactions["payment_method"] = (
    transactions["payment_method"]
    .astype(str)
    .str.upper()
    .str.strip()
)

payment_map = {
    "PHONEPE": "UPI",
    "GOOGLEPAY": "UPI",
    "GOOGLE PAY": "UPI",
    "GPAY": "UPI",

    "CC": "CREDIT CARD",
    "CREDIT_CARD": "CREDIT CARD",

    "C.O.D": "COD",
    "CASH ON DELIVERY": "COD"
}

transactions["payment_method"] = (
    transactions["payment_method"]
    .replace(payment_map)
)

transactions["payment_method"].value_counts()
```
